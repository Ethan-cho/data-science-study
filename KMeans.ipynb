```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Data Load
df = pd.DataFrame([
    [1,1,3],
    [2,0,0],
    [3,1,3],
    [4,1,3],
    [5,1,2]
], columns=['RiskX', 'RiskY', 'RiskZ'])
```

```python
model = KMeans(n_clusters=2)
model.fit(df)
```

```python
import pickle
filename = 'model.dat'

pickle.dump(model, open(filename, 'wb'))
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
result = loaded_model.predict(df)
print(result)
```

```python
y_predict = model.fit_predict(df)
#print(y_predict)
```

```python
df['cluster'] = y_predict
print(df)
#plt.plot(df)
#plt.show()
plt.plot(df['RiskX'], df['RiskY'], '^', color='red')
#plt.scatter(df['RiskX'], df['RiskY'], c=y_predict, cmap=plt.cm.Set1)

plt.show()
```
