 2009년 매출금액(SALES_AMT) 대비 2019년 매출금액(SALES_AMT)의 변화 정도를 확인하고자 한다. 다음 수식을 이용하여 증감률(%)을 기술하시오.

```python
df3 = df.copy()
```

```python
df3['YEAR_MONTH'] = pd.to_datetime(df3['YEAR_MONTH'], format='%Y-%b')
df3.head()
```

```python
df.head()
```

```python
sales_2009 = df3.loc[df3['YEAR_MONTH'].dt.year == 2009, 'SALES_AMT'].sum()
```

```python
sales_2019 = df3.loc[df3['YEAR_MONTH'].dt.year == 2019, 'SALES_AMT'].sum()
```

```python
change_rate = (sales_2019 - sales_2009) / sales_2009 * 100
```

```python
print(round(change_rate, 3))
```

 2009년과 2019년의 매출금액(SALES_AMT) 평균이 통계적으로 유의한 차이가 있는지 검정하고자 한다. 적절한 검정 수행 후, 유의 확률(P-value)의 값을 기술하시오.
단, 검정 시 세운 대립 가설(H₁)은 다음과 같다.
대립 가설(H₁): 2009년과 2019년의 매출금액의 평균이 같지 않다

```python
from scipy.stats import ttest_ind
```

```python
sales_2009 = df3.loc[df3['YEAR_MONTH'].dt.year == 2009, 'SALES_AMT']
sales_2019 = df3.loc[df3['YEAR_MONTH'].dt.year == 2019, 'SALES_AMT']
```

```python
t_stat, p_value = ttest_ind(sales_2009, sales_2019)
```

```python
print(f"p-value is {p_value}")
```

```python
round(p_value,3)
```
